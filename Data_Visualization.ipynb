##To visualize Different plots


##*Scatter Plot*

```python
plt.scatter(df['day'], df['tip'])
plt.xlabel('Day')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```

```python
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```

```python
plt.scatter(df['sex'], df['tip'])
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```

```python
plt.scatter(df['size'], df['tip'])
plt.xlabel('Size')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```

#*Bar Plot*

```python
plt.bar(df['day'], df['tip'], color = 'green')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.bar(df['sex'], df['tip'], color = 'red', linewidth = 3)
plt.xlabel('Sex')
plt.ylabel('Tip')
plt.show()
```

```python
plt.bar(df['size'], df['tip'], color = 'blue', edgecolor = 'pink')
plt.xlabel('Size')
plt.ylabel('Tip')
plt.show()
```

```python
plt.bar(df['total_bill'], df['tip'], color = 'orange', edgecolor = 'black', linewidth = .5)
plt.xlabel('Total_bill')
plt.ylabel('Tip')
plt.show()
```
